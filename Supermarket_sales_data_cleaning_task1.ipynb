```python
import pandas as pd
df = pd.read_csv('Copy of SuperMarket Analysis.csv')
df.head()
```

```python
df.info()
df.isnull().sum()  # Shows missing values per column
df.duplicated().sum()  # Shows duplicate rows
```

```python
df = df.drop_duplicates()
```

```python
# Drop rows with missing data
df = df.dropna()

# Or fill missing sales column with mean:
# df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
```

```python
df['Gender'] = df['Gender'].str.lower()
df['Product line'] = df['Product line'].str.lower()
df['City'] = df['City'].str.lower()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
df['unit_price'] = pd.to_numeric(df['unit_price'], errors='coerce')
df['tax_5%'] = pd.to_numeric(df['tax_5%'], errors='coerce')
df['gross_income'] = pd.to_numeric(df['gross_income'], errors='coerce')
df['gross_margin_percentage'] = pd.to_numeric(df['gross_margin_percentage'], errors='coerce')
```

```python
df['date'] = pd.to_datetime(df['date'], format='%m/%d/%Y', errors='coerce')
```
